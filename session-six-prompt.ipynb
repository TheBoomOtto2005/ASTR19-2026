```python
import math
import numpy as np
```

Imports functions that are necessary for the math and setup

```python
def sin_x(x):

        return math.sin(x)
```

Defines a function that returns the sin value of x

```python
def cos_x(x):

        return math.cos(x)
```

Defines a function that returns the cos value of x

```python
the_values = np.linspace(0, 2, 1000)
sin_value = [sin_x(x) for x in the_values]
cos_value = [cos_x(x) for x in the_values]
```

Tabulates and creates a table that intereats from 0 to 1000

```python
print(f"{'x':>10} {'sin(x)':>10} {'cos(x)':>10}")
print("-" * 34)
for i in range(10):
        print(f"{the_values[i]:10.4f} {sin_value[i]:10.4f} {cos_value[i]:10.4f}")
```

Makes a clean looking table header and line and produces the first 10 values for the table.
