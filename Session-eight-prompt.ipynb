```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,100)
```

```python
def sin_x(x):
    return np.sin(x)
def cos_x(x):
    return np.cos(x)
```

```python
fig, graph = plt.subplots(1,2)

graph[0].plot(x,sin_x(x))
graph[0].set_title("X vs Sin X")
graph[0].set_xlabel("X")
graph[0].set_ylabel("Sin X")

graph[1].plot(x,cos_x(x))
graph[1].set_title("X vs Cos X")
graph[1].set_xlabel("X")
graph[1].set_ylabel("Cos X")

plt.tight_layout()
plt.savefig("Cos_vs_Sin_X.pdf")
plt.show()
```
